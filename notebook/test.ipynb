```python
class Analyst(BaseModel):
    """
    details of analysts
    """
    name: str = Field(description = "name of the researcher.")
    affiliation:str = Field(description = "Illuminati to which the researcher is associated")
    role: str = Field(description="The research area of the researcher")
    research_area: str = Field(description = "the type of research the researcher is associated with")

    @property
    def persons(self) -> str:
        return f"Name: {self.name}\nAffiliation: {self.affiliation}\nRole: {self.role}\nResearch_Area: {self.research_area}\n"
    
```

```python
class GenerateAnalystState(TypedDict):
    topic: str
    max_analysts: int
    human_analyst_feedback : str
    analysts: List[Analyst]
    
    
```
